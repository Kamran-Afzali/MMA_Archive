```python
import requests
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
from datetime import datetime
import numpy as np
from googlesearch import search
import re
```

```python
def ranked_fighters():
    query1 = "wikipedia.org/wiki/Ultimate_Fighting_Championship_rankings"
    urls1=[]
    for j in search(query1, tld="co.in", num=10, stop=1, pause=2): 
        urls1.append(j)
    url1 = urls1[0]
    req1 = urllib.request.urlopen(url1)
    article1 = req1.read().decode()
    soup1 = BeautifulSoup(article1, 'html.parser')  
    tables = soup1.find_all('table', class_='wikitable')
    Y=[]
    for i in range(1,len(tables)):
        tabl=tables[i].find_all('tr')
        X=[]
        for j in range(1,len(tabl)):
            a=tabl[j].text.strip().splitlines()[2]
            X.append(a)
        Y.append(X)
    ranked=pd.DataFrame(Y).transpose()
    ranked.columns=["weight classes{}".format(i+1) for i in range(len(ranked.columns))]
    ranked.index=ranked.index+1
    return ranked
```
